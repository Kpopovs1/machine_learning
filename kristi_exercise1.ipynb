# Exercise One #

Using the dataset in the World_heights.csv file, where the heights of countries have already been averaged, answer the following series of questions:
    
    1. Pick two countries and compare and contrast their heights. 
    2. What are the average heights every year?
    3. How spread out are the heights between countries?
    4. What is the distribution of countries' heights over time?
   

```python
import pandas as pd
import matplotlib.pyplot as plt

# Problem 1. Pick two countries and compare and contrast their heights. 

data_set = pd.read_csv("world_height.csv")

# country 1: Canada

can_df = data_set[(data_set.Entity == "Canada")]
print can_df

can_df.plot(x='Year', y='height')

print('\n')

print 'Mean:', can_df.height.mean()
print 'Median', can_df.height.median()
print 'Standard Deviation:', can_df.height.std()
print 'Variance:', can_df.height.var()
```

## Canada: ##

The mean height in Canada between the years 1810 and 1980 was 173.22 cm, and the standard deviation was 3.1.
There was an obvious decline in the mean in 1900 as shown by the graph, when this last one dropped to 169.9, and then an immediate and steady incline from the years 1910 - 1980. 


```python
# country 2: Italy

ita_df = data_set[(data_set.Entity == "Italy")]
print(ita_df)

ita_df.plot(x='Year', y='height')

print('\n')

print 'Mean:', ita_df.height.mean()
print 'Median', ita_df.height.median()
print 'Standard Deviation:', ita_df.height.std()
print 'Variance:', ita_df.height.var()
```

## Italy ##

The mean height in Italy between the years 1810 to 1980 was 166.87 cm, while the standard deviation was 4.0.
There was a steady incline in the population's average height between the years 1880 - 1980, as shown by the graph.


```python
# compare and contrast the two countries:

ax = can_df.plot(x='Year', y='height')

ita_df.plot(ax=ax, x='Year', y='height')
```

## Canada & Italy

The Graph above represents Canada as a blue line and Italy as an orange line. Their heights (y-axis) are shown against the passing of time (x-axis).
Canada has a higher mean popoulation height throughout the years, 173.22 cm vs 166.87, while Italy has a larger standard deviation rate 3.1 vs 4.0. 
Both countries, despite their different and separated geographical regions, experienced a steady incline in population average height.

```python
# Problem 2: What are the average heights every year?

dictionary_mean = {}

for year in data_set.sort_values('Year').Year.unique():
        
    df_filtered = data_set[(data_set.Year == year)]
    
    print 'average height in', year, 'was', df_filtered.height.mean()
    
    dictionary_mean.update({year : df_filtered.height.mean()})

#visual representation of data

sorted_dictionary = sorted(dictionary_mean.items())
x, y = zip(*sorted_dictionary)

plt.ylabel("Height")
plt.xlabel("Years")
plt.title("Average Heights Per Year")
plt.plot(x,y)
plt.show()
```

```python
# Problem 3: How spread out are the heights between countries?

dictionary_std = {}

for country in data_set.Entity.unique():
    
    df_filtered = data_set[(data_set.Entity == country)]
    dictionary_std.update({country : df_filtered.height.std()})
    
'''   
    print country, ':'
    print 'height variance in', country, 'is', df_filtered.height.var()
    print 'height standard deviation in', country, 'is', df_filtered.height.std(), '\n'
''' 
    
# separating by continent to see if there is an interesting trend in the data?

africa_dict = pd.read_csv("africa_height.csv")

asia_dict = pd.read_csv("asia_height.csv")

na_dict = pd.read_csv("na_height.csv")

sa_dict = pd.read_csv("sa_height.csv")

eu_dict = pd.read_csv("eu_height.csv")

oceania_dict = pd.read_csv("oceania_height.csv")

continents_dict = {}

continents_dict.update({"Africa" : africa_dict.height.std()})
print 'Standard Deviation for Africa is', africa_dict.height.std()

continents_dict.update({"Asia" : asia_dict.height.std()})
print 'Standard Deviation for Asia is', asia_dict.height.std()

continents_dict.update({"North Amer." : na_dict.height.std()})
print 'Standard Deviation for North America is', na_dict.height.std()

continents_dict.update({"South Amer." : sa_dict.height.std()})
print 'Standard Deviation for South America is', sa_dict.height.std()

continents_dict.update({"E.U." : eu_dict.height.std()})
print 'Standard Deviation for Europe is', eu_dict.height.std()

continents_dict.update({"Oceania" : oceania_dict.height.std()})
print 'Standard Deviation for Oceania is', oceania_dict.height.std()


plt.ylabel("Standard Deviation")
plt.xlabel("Continents")
plt.title("Standard Deviation Per Continent")
plt.bar(range(len(continents_dict)), continents_dict.values(), 0.4, align='center')
plt.xticks(range(len(continents_dict)), list(continents_dict.keys()))
plt.show()
```

```python
# Problem 4: What is the distribution of countries' heights over time? 
# Does the spread grow larger as you move forward in the arrow of time? Or smaller?

dictionary_std = {}

for year in data_set.sort_values('Year').Year.unique():
    
    df_filtered = data_set[(data_set.Year == year)]
    print 'standard deviation in', year,'was', df_filtered.height.std()  
    dictionary_std.update({year : df_filtered.height.std()})

#visual representation of data

sorted_dictionary = sorted(dictionary_std.items())
x, y = zip(*sorted_dictionary)

plt.ylabel("Standard Deviation")
plt.xlabel("Years")
plt.title("Standard Deviation per Year")
plt.plot(x,y)
plt.show()
```

## Final Thoughts

Overall, the mean height has increased from 165.1 to 170.7 cm between the years 1810 and 1980. 
The standard deviation has also increased (see graphs above) from 2.7 to 5.2. 

For problem no.3, the average spread between countries' heights (which can be shown when the code is un-commented) is too large of a dataset to portray the spread. I separated the data between continents to better grab a visual of the spread.

I observed the following information:

    1. Countries in the continet of Oceania and North America show the largest spread between heights.
    2. Countries in Africa and South America showed the smallest standard deviation between heights.
