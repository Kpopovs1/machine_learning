$\color{Green}{\textbf{Defining Probabilistic Models:}}$ 

Probabilistic Graphical Models are frameworks for modeling variables and their complex interactions in order to make future predictions. They mathematically represent complex problems, wich might be stochastic and contain a large level of $\color{Grey}{uncertainty}$, through the use of probability distribution.

Probabilistic graphical models have a wide use in machine learning, computer vision, natural language processing, and medicine, and combine two essential mathematical fields: probability and graph theory. 
Check out the following Link to learn more about real-world applications of Probabilistic Graphical Models: https://ermongroup.github.io/cs228-notes/preliminaries/applications/

Since the world contains exponentially-sized uncertainty, one method to simplify things is to ussume that their probabilities are conditionally independent. This method of simplification is called $\color{Grey}{Naive \space Bayes \space assumption}$, which assumes that the existence of a feature is unrelated to the existence of other features. 

$$ P(y,x_1,x_2,...,x_n) = p(y) \prod_{i=1}^{n} p(x_i|y)$$


In the graph above, the vertices represent random variables and the arrows represent their probabilistic relationships. 

This naive method assumes that all features are independent, which is not always the case. There are other models to represent probability distributions that might work better. In order to understand how to come up with such models, we will explore different types such as Bayesian Networks (directed graphs) and Markov Random Fields (undirected graphs) and their properties.

## Bayesian Networks
